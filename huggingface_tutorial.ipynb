```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

* The pipeline("sentiment-analysis") automatically loads a pre-trained sentiment analysis model (usually distilbert-base-uncased-finetuned-sst-2-english by default).
* It processes text input and returns whether the sentiment is positive or negative along with a confidence score.

```python
classifier("I am learning AI and this is my passion")
```

```python
classifier(" I do not trust anyone")
```

```python
generation = pipeline("text-generation")
```

```python
generation("I love to learn Data Science, but it takes time")
```

```python
summarizer = pipeline("summarization")
```

```python
text = "The lion (Panthera leo) is a large cat of the genus Panthera, native to Africa and India. It has a muscular, broad-chested body; a short, rounded head; round ears; and a dark, hairy tuft at the tip of its tail. It is sexually dimorphic; adult male lions are larger than females and have a prominent mane. It is a social species, forming groups called prides. A lion's pride consists of a few adult males, related females, and cubs. Groups of female lions usually hunt together, preying mostly on medium-sized and large ungulates. The lion is an apex and keystone predator."
```

```python
print(summarizer(text, max_length = 20))
```

```python
classifier = pipeline("zero-shot-classification")
res = classifier(
    "I want to become a prime minister of India",
    candidate_labels = ["politics", "education", "business"]
)
```

```python
print(res)
```

```python
# Load Custom Model
summarizer2 = pipeline("summarization", model = "google-bert/bert-base-uncased")
```

```python
text
```

```python
summarizer2(text, max_new_tokens = 20)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
# Loading tokenizer of t5 model from google(name copied from Huggingface)
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
# Model is getting loaded
model = AutoModelForSequenceClassification.from_pretrained("google/flan-t5-large")
```

```python
tokenizer
```

```python
model
```

```python
# Load model and save to our infrastructure so that we can load in the future
# Create directory
save_directory = "my_model_dir"
```

```python
tokenizer.save_pretrained(save_directory)
model.save_pretrained(save_directory)
```
